### Popularity Based Recommender System

```python
df = ratings.merge(books, on = 'ISBN')
```

```python
df.shape
```

```python
df.head()
```

```python
df_rating = df.groupby("Book-Title").count()['Book-Rating'].reset_index()
df_rating.rename(columns = {'Book-Rating':'Num_rating'}, inplace = True)
df_rating.head()
```

```python
df_rating_avg = df.groupby("Book-Title").mean()['Book-Rating'].reset_index()
df_rating_avg.rename(columns = {'Book-Rating':'Num_rating_avg'}, inplace = True)
df_rating_avg.head()
```

```python
df_popular = df_rating.merge(df_rating_avg, on = 'Book-Title')
df_popular
```

```python
df_popular = df_popular[df_popular['Num_rating']>=250].sort_values("Num_rating_avg", ascending = False).head(50)
```

```python
df_popular['Num_rating_avg'] = np.round(df_popular['Num_rating_avg'])
```

```python
df_popular = df_popular.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')
```

```python
df_popular.keys()
```

```python
df_popular = df_popular[['Book-Title','Book-Author', 'Image-URL-M','Num_rating', 'Num_rating_avg']]
```

```python
df_popular.shape
```

### Collaborative Filtering

```python
x = df.groupby("User-ID").count()['Book-Rating'] > 200
users = x[x].index
```

```python
filter_rating = df[df['User-ID'].isin(users)]
```

```python
filter_rating
```

```python
y = df.groupby("Book-Title").count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_rating = df[df['Book-Title'].isin(famous_books)]
final_rating
```

```python
final_rating.duplicated().sum()
```

```python
pt = final_rating.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace = True)
```

```python
pt = pt.iloc[:1000, :10000]
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
similarity_scores[0]
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x:x[1], reverse = True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
